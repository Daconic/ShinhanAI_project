## 라이브러리 불러오기

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import FinanceDataReader as fdr
```

## 코스피 상위 30개 종목 불러오기

```python
top30_stock = fdr.StockListing('KOSPI')[:30][['Code', 'Name']]
top30_stock
```

## 네이버 뉴스 크롤링 함수 불러오기

```python
def crawler(company_code, maxpage):
    
    result = pd.DataFrame(columns=["날짜", "기사제목"])
    
    for page in range(1, int(maxpage)+1):
        url = 'https://finance.naver.com/item/news_news.nhn?code=' + str(company_code) + '&page=' + str(page)
        source_code = requests.get(url).text
        html = BeautifulSoup(source_code, "lxml")
        
        # 뉴스 제목 
        titles = html.select('.title')
        title_result = [re.sub('\n', '', title.get_text()) for title in titles]
         
        # 뉴스 날짜 
        dates = html.select('.date') 
        date_result = [date.get_text() for date in dates] 
                
        # 변수들을 데이터프레임에 추가하기 
        df_result = pd.DataFrame({"날짜": date_result, "기사제목": title_result})
        result = pd.concat([result, df_result], ignore_index=True)
        print(f"다운로드 중... {page}/{maxpage}")
    
    return result
```

```python
# 상위 30개 종목 이름 리스트화
top_30_names  = list(top30_stock['Name'])
```

```python
# 대략 27분 걸림
import time

start_time = time.time()

crawled_news = {}

for i, code in enumerate(top30_stock['Code']):
    crawled_news[top_30_names[i]] = crawler(code, 500)
    
end_time = time.time()

print("총 실행 시간: ", end_time - start_time, "초")

```

```python
backup = crawled_news.copy()
```

```python
crawled_news = backup.copy()
```

```python
for name in top_30_names:

    제목내용_특수문자제거 = []
    for i in crawled_news[name]['기사제목']:
        w = re.sub(r"[^\w\s]", "", i)
        제목내용_특수문자제거.append(w)
    crawled_news[name]['기사제목'] = 제목내용_특수문자제거
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")

model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC")
```

```python
# 감정 점수 출력 
import time
from transformers import pipeline

classifier = pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
```

## 삼성바이오로직스

```python
sentiment = []
for i in crawled_news['삼성바이오로직스']['기사제목']:
    a = classifier(i)
    sentiment.append(a)
```

```python
df = pd.DataFrame({'sentiment': [item[0]['label'] for item in sentiment],
                   'score': [item[0]['score'] for item in sentiment]})
```

```python
crawled_news['삼성바이오로직스']['감정상태'] = df['sentiment']

sentiment_mapping = {'neutral': 0, 'positive': 1, 'negative': 2}
crawled_news['삼성바이오로직스']['감정상태'] = crawled_news['삼성바이오로직스']['감정상태'].apply(lambda x: sentiment_mapping[x])

crawled_news['삼성바이오로직스'] = crawled_news['삼성바이오로직스'].drop(columns=['기사제목'])
crawled_news['삼성바이오로직스']['날짜'] = pd.to_datetime(crawled_news['삼성바이오로직스']['날짜'], format='%Y.%m.%d %H:%M')
crawled_news['삼성바이오로직스']['날짜'] = crawled_news['삼성바이오로직스']['날짜'].dt.strftime('%Y-%m-%d')
crawled_news['삼성바이오로직스'].set_index(pd.to_datetime(crawled_news['삼성바이오로직스']['날짜']), inplace=True)
crawled_news['삼성바이오로직스'] = crawled_news['삼성바이오로직스'].groupby(crawled_news['삼성바이오로직스'].index).mean()
crawled_news['삼성바이오로직스']['감정상태'] = np.round(crawled_news['삼성바이오로직스']['감정상태'])
crawled_news['삼성바이오로직스'] = crawled_news['삼성바이오로직스'].resample('D').ffill()

# csv 파일로 저장 
crawled_news['삼성바이오로직스'].to_csv('Sentiment_Analysis/삼성바이오로직스_감정점수.csv')
```
