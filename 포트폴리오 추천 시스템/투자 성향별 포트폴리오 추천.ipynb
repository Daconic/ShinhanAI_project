투자 형태에 대한 자산 배분 비율
성향 분석이 아닌 개개인의 선택에 맞추어 진단한다다
*적금의 경우 적금과 주택청약의 비율을 8:2로 지정한다
1. 안정적인 적금형 = {적금:100}
2. 투자 맛보기형 = {적금:80, 안정적 투자:20}
3. 투자 즐기기형: = {적금:60, 안정적 투자:40}
4. 도전 즐기기형형 = {적금:60, 공격적 투자:40}

```python
import matplotlib.pyplot as plt
def pie_chart():

  #num(투자 번호)에 따른 투자 포트폴리오 반환
  num = int(input('1. 안정적인 적금형 = {적금:100}\n 2. 투자 맛보기형 = {적금:80, 안정적 투자:20}\n 3. 투자 즐기기형: = {적금:60, 안정적 투자:40}\n 4. 도전 즐기기형형 = {적금:60, 공격적 투자:40}\n번호 입력'))
  if num == 1:
    ratio = [80, 20]
    labels = ['적금', '주택청약']
    colors = ['silver', 'gold']

    plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors = colors,shadow=True)
    return plt.show();

  if num == 2:
    ratio = [64, 16, 20]
    labels = ['적금', '주택청약', '투자']
    explode = [0.01, 0.01, 0.07]
    colors = ['silver', 'gold', 'whitesmoke']

    plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode = explode, colors = colors, shadow=True)
    return plt.show();    

  else:
    ratio = [48, 12, 40]
    labels = ['적금', '주택청약', '투자']
    explode = [0.01, 0.01, 0.07]
    colors = ['silver', 'gold', 'whitesmoke']

    plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode = explode, colors = colors, shadow=True)
    return plt.show();
```

```python
def 설문():
  ans = input('투자 형태가 안정적인 적금형 {적금: 100}인가요? y/n')
  if (ans == 'n') or (ans == 'N'):
    
    q1 = int(input('1. 고객님의 총 자산 규모(순자산)은 다음 중 어느 수준입니까?\n (1)	10만원 미만\n (2)	10만원 이상 ~ 70만원 미만\n (3)	70만만원 이상 ~ 130만원 미만\n (4)	130만원 이상 ~ 200만원 미만\n (5)	200만원 이상\n'))
    q2 = int(input('2. 고객님의 연간 소득액은 어떻게 되십니까?(대략 세뱃돈 포함하여)\n (1)	50만원 미만\n (2)	50만원 이상 ~ 90만원 미만\n (3)	90만만원 이상 ~ 130만원 미만\n (4)	130만원 이상 ~ 170만원 미만\n (5)	200만원 이상\n'))
    #'https://www.mk.co.kr/news/economy/10723255' 이 기사를 참고하여 q2의 항목을 지정하였음
    q3 = int(input('3. 고객님이 선택하신 유형별 비중은 어떻게 되십니까?\n (1) 투자 맛보기형 = {적금:80, 안정적 투자:20} \n (2) 투자 즐기기형: = {적금:60, 안정적 투자:40} \n (3) 도전 즐기기형 = {적금:60, 공격적 투자:40}\n'))
    q4 = int(input('4. 투자해보신 경험이 있나요? 있다면 기간은?\n (1) 있음, 1년 미만\n (2) 있음, 1년 이상 ~ 3년미만\n (3) 있음, 3년 이상\n (4) 없음\n'))
    q5 = int(input('5. 고객님의 금융지식 수준(이해도)는 어느정도라고 생각하십니까?\n (1)	금융투자상품에 투자해 본 경험이 없음\n (2)	금융투자상품(주식 , 채권 및 펀드 등)의 구조 및 위험을 일정 부분 이해\n (3)	금융투자상품(주식 , 채권 및 펀드 등)의 구조 및 위험을 깊이 있게 이해\n'))
    q6 = int(input('6. 고객님의 금융투자상품 취득 및 처분 목적은 무엇입니까?\n (1)	자산증식\n (2)	교육과 경험\n (3)	미래자금\n'))
    q7 = int(input('7. 고객님은 다음 중 어떤 목적으로 투자하는 편입니까?\n (1)	투자 수익을 고려하나 원금 보존이 더 중요\n (2)	원금 보존을 고려하나 투자 수익이 더 중요\n (3)	손실 위험이 있더라도 투자 수익이 중요\n'))
    q8 = int(input('8. 고객님께서 투자하고자 하는 자금의 투자예정 기간은 얼마나 되십니까?\n (1)	3년 미만\n (2)	3년 이상 ~ 5년 미만\n (3)	5년 이상 ~ 7년 미만\n (4)	7년 이상 ~ 10년 미만\n (5)	10년 이상\n'))
    q9 = int(input('9. 고객님께서 금융상품 투자를 통해 기대하는 수익과 감수할 수 있는 손실을 가장 잘 표현한 것은 어떤 것입니까?\n (1)	위험도를 낮춰 시중 금리만큼의 수익\n (2)	원금의 일부 손실을 감수 및 시중금리보다 조금 높은 수익\n (3)	원금 손실을 감수하여 좀 더 높은 수익을 기대\n (4)	원금을 많이 잃더라도 더 높은 수익 추구\n'))


    #설문지를 바탕으로 투자자 성향 점수내기
    #q1과 q2는 q1을 q2로 나눈 값을 다섯 단계로 나누어 점수 배분분 
    sum = 0
    if q1/q2 <0.25:
      sum += 1
    elif q1/q2 <0.5:
      sum += 2
    elif q1/q2 <0.75:
      sum += 3
    else: 
      sum += 4

    #q3
    if q3 == 1:
      sum += 0
    elif q3 == 2:
      sum += 2
    else:
      sum += 4 

    #q4
    if q4 == 1:
      sum += 1
    elif q4 == 2:
      sum += 2
    elif q4 == 3:
      sum += 3
    else:
      sum += 0

    #q5
    if q5 == 1:
      sum += 1
    elif q5 == 2:
      sum += 2
    else:
      sum += 3

    #q6
    if q6 == 1:
      sum += 4
    elif q6 == 2:
      sum += 0
    else:
      sum += 2

    #q7
    if q7 == 1:
      sum += 1.3
    elif q7 == 2:
      sum += 2.6
    else:
      sum += 4

    #q8 
    if q8 == 1:
      sum += 0.8
    elif q8 == 2:
      sum += 1.6
    elif q8 == 3:
      sum += 2.4
    elif q8 == 4:
      sum += 3.2
    else:
      sum += 4 

    #q9
    if q8 == 1:
      sum += 0
    elif q8 == 2:
      sum += 1
    elif q8 == 3:
      sum += 2
    elif q8 == 4:
      sum += 3
    else:
      sum += 4

    sum = round(sum, 2)
    if sum < 10:
      투자형태 = '안정형'
    elif sum < 20:
      투자형태 = '위험중립형'
    else:
      투자형태 = '공격투자형'
    print('\n\n당신의 점수는 {}점이며, 투자자형태는 {}입니다'.format(sum, 투자형태))

    return 투자형태

  else:
    print('적금형은 투자자 성향 분석이 필요 없습니다!')
```

```python
def portfolio(df, 투자형태):
  df['safe_grade'] = 0
  df.loc[df['Sharpe Ratio'] >= 3, 'safe_grade'] = 3
  df.loc[df['Sharpe Ratio'] < 3, 'safe_grade'] = 2
  df.loc[df['Sharpe Ratio'] < 2, 'safe_grade'] = 1


  if 투자형태 == '위험중립형':
    print(df.loc[df['safe_grade'] == 2, 'Portfolio'])
  elif 투자형태 == '안정형':
    print(df.loc[df['safe_grade'] == 3, 'Portfolio'])
  else:
    print(df.loc[df['safe_grade'] == 1, 'Portfolio'])
```
