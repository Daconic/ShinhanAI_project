```python
import pandas as pd
import numpy as np
import folium
import time

from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.keys import Keys
```

```python
# 크롬 드라이버 경로
chromedriver_path = "chromedriver.exe"

# 크롬 옵션 설정
chrome_options = webdriver.ChromeOptions()

# Service 객체 생성
service = Service(chromedriver_path)

# 크롬 드라이버 초기화
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.implicitly_wait(10)

url = "https://www.shinhancard.com/pconts/html/card/check/MOBFM282R11.html"
driver.get(url)
```

```python
# 윈도우 화면 최대화
driver.maximize_window()
```

```python
# 페이지 전체 HTML 파싱
html = driver.page_source
soup = BeautifulSoup(html,"html.parser")
```

```python
container = soup.find("div",class_="tab_wrap")
tbody = container.find("div",class_="card_list_common gap60_30")
li_all = tbody.find_all("li")
```

```python
name_list = []
bold_list = []
common_list = []
description_list = []
one_list = []
two_list = []
three_list = []

for idx, i in enumerate(li_all):
    bold_name = ''
    
    print(idx)
    name = i.find("a", class_ = "card_name").text
    print(name)
    name_list.append(name)

    searchbutton = driver.find_element(By.CSS_SELECTOR, f"#section01_1 > div > div > ul > li:nth-child({idx+1}) > div.card_name_wrap > a")
    searchbutton.send_keys(Keys.ENTER)
    
    html = driver.page_source
    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#cardCompareAfter > div.card_detail.gap80_40 > div.card_name_wrap > p')))
    
    soup = BeautifulSoup(html,"html.parser")
    
    description = driver.find_element(By.CSS_SELECTOR, '#cardCompareAfter > div.card_detail.gap80_40 > div.card_name_wrap > p').text
    print(description)
    description_list.append(description)
    
    one_list.append(driver.find_element(By.CSS_SELECTOR, f"#cardCompareAfter > div.card_detail.gap80_40 > div.right_wrap > ul > li:nth-child(1)").text) 
    two_list.append(driver.find_element(By.CSS_SELECTOR, f"#cardCompareAfter > div.card_detail.gap80_40 > div.right_wrap > ul > li:nth-child(2)").text) 
    
    print(driver.find_element(By.CSS_SELECTOR, f"#cardCompareAfter > div.card_detail.gap80_40 > div.right_wrap > ul > li:nth-child(1)").text)
    print(driver.find_element(By.CSS_SELECTOR, f"#cardCompareAfter > div.card_detail.gap80_40 > div.right_wrap > ul > li:nth-child(2)").text)
    
    try:
        a = driver.find_element(By.CSS_SELECTOR, f"#cardCompareAfter > div.card_detail.gap80_40 > div.right_wrap > ul > li:nth-child(3)").text
    except:
        a = '없음'
        
    print(a)
    three_list.append(a)    
        
    content = soup.find('ul', class_='all_view')
    
    for j in range(len(content.find_all('a', class_='btn_move'))):
        bold = driver.find_element(By.CSS_SELECTOR, f"#section08_0 > div > div > div > ul > li:nth-child({j+1}) > a > div > strong").text
        bold_name += bold + ' '
    bold_list.append(bold_name)

    #뒤로가기
    time.sleep(1)
    driver.back()
```

```python
df = pd.DataFrame()
df['카드명'] = name_list
df['설명'] = description_list
df['혜택1'] = one_list
df['혜택2'] = two_list
df['혜택3'] = three_list
df['세부혜택'] = bold_list
df.head()
```

```python
# 청소년에게 부적합한 카드 제외
df_total = df[~df['카드명'].str.contains('Finnq|쇼핑|마트|EV|더본|신세계|증권|유치원|대덕|가족|나주|성남|공무원|스포츠|Change-Up|주유소|SOCAR|한국야쿠르트|더본|몰테일|홈플러스|경차|택시|화물|장애인|경차|아이|하이패스|주거래|스카이패스|아시아나|Global')].reset_index(drop=True)
df_total.tail()
```

```python
df_total['설명'] = df_total['설명'].str.replace('\n', ' ')
df_total['혜택1'] = df_total['혜택1'].str.replace('\n', ' ')
df_total['혜택2'] = df_total['혜택2'].str.replace('\n', ' ')
df_total['혜택3'] = df_total['혜택3'].str.replace('\n', ' ')
df_total.head()
```
