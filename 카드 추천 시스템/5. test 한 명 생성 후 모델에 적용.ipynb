```python
from faker import Faker
import random
import pandas as pd

fake = Faker('ko-KR')
```

```python
name = [fake.name() for i in range(1)]
age = [fake.pyint(min_value=8, max_value=19) for i in range(1)]
```

```python
k = 1
hobby = ['스포츠', '영화관람', '뷰티', '쇼핑', '도서', '게임', '여행', '카페', '패션']
random_hobbies= random.choices(hobby, k=k)

gender = ['남', '여']
random_gender= random.choices(gender, k=k)

benefit = ['대중교통', '야구', '도서', '뷰티', '배달', '카페', '쇼핑', '영화관람', '놀이동산', '넷플릭스', '편의점', '통신비']
random_benefit= random.choices(benefit, k=k)
```

```python
id_list = []
for i in range(10001, 10002):
    id_number = f'a{i:05}'
    id_list.append(id_number)
```

```python
transport = []
transport_num = []

con = []
con_num = []

delivery  = []
delivery_num = []

cafe = []
cafe_num = []

ott = []
ott_num = []

movie = []
movie_num = []

shopping = []
shopping_num = []

book = []
book_num = []

cosmetics = []
cosmetics_num = []

baseball = []
baseball_num = []

park = []
park_num = []

phone = []
phone_num = []

for _ in range(1, k+1):
    random_transport = random.randint(300, 720) // 100 * 100 # 한 번 탑승시 요금
    transport.append(random_transport)
    transport_cnt = random.randint(5, 70)
    transport_num.append(transport_cnt)
    
    random_con = random.randint(2000, 50000) // 100 * 100
    con.append(random_con)
    con_cnt = random.randint(2, 30)
    con_num.append(con_cnt)
    
    random_delivery = random.randint(10000, 30000) // 100 * 100
    delivery.append(random_delivery)
    delivery_cnt = random.randint(2, 10)
    delivery_num.append(delivery_cnt)
    
    random_cafe = random.randint(4500, 20000) // 100 * 100
    cafe.append(random_cafe)
    cafe_cnt = random.randint(2, 20)
    cafe_num.append(cafe_cnt)
    
    random_ott = random.randint(2500, 5000) // 100 * 100
    ott.append(random_ott)
    ott_cnt = random.randint(0, 6)
    ott_num.append(ott_cnt)
    
    random_movie = random.randint(13000, 25000) // 100 * 100
    movie.append(random_movie)
    movie_cnt = random.randint(0, 3)
    movie_num.append(movie_cnt)
    
    random_shopping = random.randint(10000, 50000) // 100 * 100
    shopping.append(random_shopping)
    shopping_cnt = random.randint(0, 5)
    shopping_num.append(shopping_cnt)
    
    random_book = random.randint(18000, 30000) // 100 * 100
    book.append(random_book)
    book_cnt = random.randint(0, 5)
    book_num.append(book_cnt)
    
    random_cosmetics = random.randint(10000, 50000) // 100 * 100
    cosmetics.append(random_cosmetics)
    cosmetics_cnt = random.randint(0, 5)
    cosmetics_num.append(cosmetics_cnt)
    
    random_baseball = random.randint(8000, 25000) // 100 * 100
    baseball.append(random_baseball)
    baseball_cnt = random.randint(0, 2)
    baseball_num.append(baseball_cnt)
    
    random_park = random.randint(20000, 50000) // 100 * 100
    park.append(random_park)
    park_cnt = random.randint(0, 2)
    park_num.append(park_cnt)
    
    random_phone = random.randint(10000, 75000) // 100 * 100
    phone.append(random_phone)
    phone_cnt = 1
    phone_num.append(phone_cnt)
```

```python
df = pd.DataFrame()
df['이름'] = name
df['고유번호'] = id_list
df['관심사'] = random_hobbies
df['나이'] = age
df['성별'] = random_gender
df['혜택'] = random_benefit

kor = ['교통', '편의점', '배달', '카페', 'ott', '영화', '쇼핑', '도서', '뷰티', '야구', '놀이공원/공연', '통신']
eng = [transport, con, delivery, cafe, ott, movie, shopping, book, cosmetics, baseball, park, phone]
cnt = [transport_num, con_num, delivery_num, cafe_num, ott_num, movie_num, shopping_num, book_num, cosmetics_num, baseball_num, park_num, phone_num]

for i, j, k in list(zip(kor, eng, cnt)):
    df[f'{i}비'] = j
    df[f'{i}횟수'] = k
    df[f'{i}총비용'] = df[f'{i}비'] * df[f'{i}횟수']
```

```python
df_card = pd.read_csv('card.csv')
```

```python
card_list = df_card[df_card['Cluster'] == y_pred[0]]
```

```python
# user 각 행에서 가장 큰 값의 열 이름 찾기
max_cols = df.iloc[:,3:].apply(lambda x: x.nlargest(3).keys(), axis=1)

# 상위 3개의 열 이름 추출하고 리스트로 변환
top_3_cols = max_cols.apply(lambda x: x[:3].tolist())

cnt_list = []
for j in range(len(card_list)):
    cnt = 0
    for i in range(3):
        a = top_3_cols[0][i].split('총')[0]

        if card_list.iloc[j][a] == 1.0:
            cnt += 1
    cnt_list.append(cnt)
```

```python
card_list['cnt'] = cnt_list
card_list
```

```python
# 유사도 분석
import pandas as pd
from faker import Faker
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# tfidf 벡터화
tfidf = TfidfVectorizer()
card_tfidf = tfidf.fit_transform(card_list[['혜택1', '혜택2', '혜택3', '세부혜택']].apply(lambda x: ' '.join(x), axis=1))

df_tfidf = tfidf.transform(df[['관심사', '혜택']].apply(lambda x: ' '.join(x), axis=1))

# 코사인 유사도 계산
cos_sim = cosine_similarity(card_tfidf, df_tfidf)

# 코사인 유사도 컬럼 생성
card_list['cos'] = 0.01 + np.round(cos_sim, 3)
card_list['total'] = (card_list['cnt'] + 0.01) * card_list['cos']

# total이 높은 순으로 정렬
card_list = card_list.sort_values(['total'], ascending=False)

# 추천 결과 출력
recommendations = card_list.iloc[:3, :]

card_names = recommendations['카드명'].tolist()
card_benefits1 = recommendations['혜택1'].tolist()
card_benefits2 = recommendations['혜택2'].tolist()
card_benefits3 = recommendations['혜택3'].tolist()

print("고객님의 추천 카드는 다음과 같습니다.")
for i in range(len(card_names)):
    print(f"{i+1}. {card_names[i]} : {card_benefits1[i]}, {card_benefits2[i]}, {card_benefits3[i]}")
```
